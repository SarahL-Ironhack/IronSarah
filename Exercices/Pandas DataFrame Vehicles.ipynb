```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_excel(r"C:\Users\Sarah\Downloads\vehicles\vehicles.xlsx")
```

```python
cars.head()
```

```python
cars.loc[cars['Transmission'].str.startswith('A'), 'TransType'] = 'Automatic' #for columns transmission elements that start with A, add a new column Transtype with data Automatic
cars.head()
```

```python
cars.loc[cars['Transmission'].str.startswith('M'), 'TransType'] = 'Manual'
cars.head()
```

```python
cars['Transmission']
```

```python
avg_mpg = cars.groupby('Make', as_index=False).agg({'Combined MPG':'mean'}) #create new dataframe with aggregated DataFrame
avg_mpg.columns = ['Make', 'Avg_MPG']
avg_mpg
```

```python
data = pd.merge(cars, avg_mpg, on='Make') #merging the two data frames on column 'Make', columns Avg_MPG is at the end
data.head(10)
```

```python
data2 = cars.groupby('Make', as_index=False).agg({'Combined MPG':'sum'}) #create new DF with make and SUM
```

```python
data2
```

```python
data3 = pd.merge(cars, data2, on='Make') #merge new Data Frame with the first table
```

```python
mpg_labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
bins = pd.cut(data['Combined MPG'],5, labels=mpg_labels) #cut converts to "bins" aka groups of data, here in 5 groups according to data (automatically sorted)
bins.head(10)
```

```python
cars['bins1'] = pd.cut(data['Combined MPG'],5, labels=mpg_labels)
cars.head(10)
```

```python
cars['bins2'] = pd.qcut(data['Combined MPG'],5, labels=mpg_labels) #quantile cut, so according tu the number or values
cars.head(10)
```

```python
cutoffs = [7,14,21,23,30,40]
cars['bins3'] = pd.cut(data['Combined MPG'],cutoffs, labels=mpg_labels) #bins are predefined 
cars.head(10)
```

```python
cars['bins1'].value_counts()
```
