```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

#Seaborn is based on Matplotlib, so it is nice to import both libraries abs
```

```python
# generate the data, here as a DataFrame with 2 columns x and y 
data = pd.DataFrame()
data['X'] = np.arange(0,10,1)
data['Y'] = data['X']**2
```

```python
sns.lineplot(data.X, data.Y) 
```

```python
sns.lmplot("X", "Y", data=data, fit_reg=False)
```

```python
data = pd.read_csv(r'C:\Users\Sarah\Downloads\vehicles.csv')
sns.scatterplot(x="CO2 Emission Grams/Mile", y="Highway MPG", hue = "Drivetrain", data=data)
#library sns, plot is scatterplot, what variable for x, what variable for y, 'hue' is the variable under which we group the plots(so same modality will be same colour) 
```
