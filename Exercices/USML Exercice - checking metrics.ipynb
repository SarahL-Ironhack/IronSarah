```python
from sklearn import metrics
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

import pandas as pd
import numpy as np 
```

```python
# load dataset and turn it into dataframe
iris = datasets.load_iris()
iris_pd = pd.DataFrame(iris['data'],columns =['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'] )
y_true=iris['target']
```

Exercise :
KMeans, DBSCAN and Hierarchical clustering, checking the variants with 2,3,4,5 clusters (where it is possible)


```python
# implement kmeans, HC, DBSCAN
output_dict={}

#Here Kmeans and Hierarchical classifier
for i in range(2,5):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(iris_pd)
    key ="kmeans"+str(i)
    output_dict[key] = kmeans.labels_ # add something to store the labels
    hc = AgglomerativeClustering(n_clusters=i)
    hc.fit_predict(iris_pd)
    key ="hc"+str(i)
    output_dict[key] = hc.labels_
    # HC =KMeans(n_clusers=i)
    
eps =[0.02, 0.15,0.18,0.1] #values found here under with DBSCAN
min_samples = [2, 4, 4, 2] #values found here under with DBSCAN

for i,j in zip(eps, min_samples) :
    db = DBSCAN(eps=i, min_samples=j).fit(iris_pd)
    key2="db"+str(i)
    output_dict[key2] = db.labels_

output_dict #output of all three models in one dictionary 

# now we have three parameters to work with
```

Create one function that allows to estimate the following list of metrics for the unsupervised learning.
[Homogeneity, Completeness, V-measure, Silhouette Coefficient, Variance Ratio Criterion]:

```python
from sklearn.metrics.cluster import homogeneity_score
from sklearn.metrics.cluster import completeness_score
from sklearn.metrics.cluster import v_measure_score
from sklearn.metrics.cluster import silhouette_score
from sklearn.metrics.cluster import calinski_harabasz_score
from sklearn.metrics.cluster import adjusted_mutual_info_score
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import fowlkes_mallows_score
```

```python
def metrics_testing (X, y_true, y_pred):
    homogenity = homogeneity_score(y_true, y_pred)
    print('homogenity is', homogenity) #how points are similar to each other. point inside cluster most homogenous 
    # we see score gets good for 3 or 4 clusters , looking for min number of clusters, with best score
    
    completeness = completeness_score(y_true, y_pred)
    print('completeness is', completeness)
    # if all points of data observations are from same type, and classified as same cluster
    # very similar to confusion matrix : is all predicted in good way or not ? 
    # here good for 2 and 3
    
    v_measure = v_measure_score(y_true, y_pred)
    print('v_measure is', v_measure)
    # based on homogenity and completeness v = (1 + beta) * homogeneity * completeness / (beta * homogeneity + completeness)
    # higher measure give better result
    # here good for 3 
    
    adjusted_rand = (adjusted_rand_score(y_true,y_pred))
    print('adjusted_rand is', adjusted_rand)
    # belongs to one cluster or not. Has is been well classified ? 
    # if adjusted rand is near 0 -> means we have some kind of random label, close to 1 -> vers similar 
    # if 1 : ypred and ytrue give same labels
    # here good for 3 or 4
    
    silhouette = silhouette_score(X, y_pred)
    print('silhouette_score is', silhouette)
    # simplest one
    # distance between elements 
    #bigger coef is better
    
    variance_ratio = calinski_harabasz_score(X, y_pred)
    print('variance ratio is', variance_ratio)
    # 
    
    adjusted_mutual = adjusted_mutual_info_score(y_true,y_pred)
    print('adjusted mutual is', adjusted_mutual)
    # how many information is shared between clusted. There should not be a lot of information shared
    # result is non-positive scores
    # make sure you have no negative scores
    
    fowlkes_mallow = fowlkes_mallows_score(y_true,y_pred)
    print("\nFowlkes Mallow is:",fowlkes_mallow )

for key in output_dict:
    print('\n',key)
    metrics_testing(X=iris_pd, y_true=iris['target'], y_pred=output_dict[key])
```

```python
eps_d=[]
for eps in range(1,20):
    for min_samples in range(2,20):
        db = DBSCAN(eps=eps/100, min_samples=min_samples).fit(iris_pd) #eps1 , eps is distance between two samples
        cl = len(np.unique(db.labels_))
        eps_d.append([eps/100,min_samples,cl])
df_db= pd.DataFrame(eps_d, columns=['e','ms','cl'])
df_db['cl'].value_counts() # 2 clusters seems interesting, so digging this 
```
