```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x1 = np.arange(0,10,1) #takes everything from 0 to 10 with step 1
x2 = np.linspace(0,5,10) #from 0 to 5 , evenly spaced over interval of 10
```

```python
plt.plot(x1, x1, label = 'linear') #plot two lines, one labeled linear, the second quadratic
plt.plot(x1,x2**2, label ='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show
```

```python
fig, [ax1, ax2] = plt.subplots(1,2) # figures, 2 subplots (ax1 and ax2) , defined what is subplot 1 and subplot2 (1,2) = (rows,columns)

#plot 1
ax1.plot(x1,x2) #ax1 as defined in first line
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')   

#plot 2
ax2.plot(x1,np.cos(x1))
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

plt.show()
```

```python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
fig, ax = plt.subplots()

ax.imshow(Z)

plt.show()
```

# PLOTTING MULTIPLE DATA SERIES

```python
import pandas as pd
vehicles = pd.read_csv(r'C:\Users\Sarah\Downloads\vehicles.csv')
vehicles.columns
```

```python
vehicles.sort_values(by=["CO2 Emission Grams/Mile", "City MPG", "Highway MPG"],inplace=True)
# sorting my dataframe by several variables , inplace=True to modify initial data 
```

```python
vehicles.plot(x="CO2 Emission Grams/Mile", y=["City MPG", "Highway MPG"])
# one X axis and two y varaibles to show the tendancies 
```

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Highway MPG")
#scatter does here not accept two variables
```

```python
vehicles_mean = vehicles[["Highway MPG", "City MPG", "Drivetrain"]].groupby(["Drivetrain"]).agg("mean")
#preparing new DF that will have aggregated data : vehicles_mean
vehicles_mean.plot.barh(color = {"purple","grey"}) #bar for vertical bars, barh for horizontal bars, in () color for example
```
