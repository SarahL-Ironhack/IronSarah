```python
# copy paste the request for a dataFrame, for european region, for 3 countries, time range from 2003 to 2019 
business = wb.data.DataFrame('IC.REG.DURS', wb.region.members('ECS'))
business
```

```python
ren = wb.data.DataFrame('IC.REG.DURS', ['FRA','UKR','EST'], time=range(2003,2019))
ren.plot.bar() #data / plot / bar
```

```python
wb.data.DataFrame('IC.REG.DURS', ['FRA','UKR','EST'], time=range(2003,2019)).T.plot()
ren.T.plot()
```

```python
g5 = business.sort_values(by=['YR2019'],ascending=True)[:3] # sort values for Yr2019 , last 3 countries
ax = business.T.plot(color='lightgray', legend=False) # 
g5.T.plot(ax=ax, figsize=(15,5))
```
