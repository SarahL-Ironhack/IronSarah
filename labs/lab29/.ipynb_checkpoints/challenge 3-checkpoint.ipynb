**First, identify two groups of customers from the data set.** The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). ## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# aggregate amount_spent for unique customers => use CustomerID 
orders = orders[['CustomerID','amount_spent','Country']]

unique_cust = orders.groupby('CustomerID',as_index=False).sum()
unique_cust.sort_values(by='amount_spent', ascending = False)
```

```python
### Find customers in given percentile 
vip_perc = unique_cust.amount_spent.quantile(0.95)
# result is 5840,18 which is the amount of the 95th percentile 
VIP = unique_cust.loc[(unique_cust['amount_spent'] >= vip_perc)]
VIP
```

```python
# identify preferred customers 
pref_cust_perc = unique_cust.amount_spent.quantile([.75, .95])
pref_cust_perc # 0.75 is 1661.640, 0.95 is 5840.182

pref_cust = unique_cust.loc[(unique_cust['amount_spent'] >= 1661.640) & (unique_cust['amount_spent'] < 5840.182)]
pref_cust 
```

```python
#Add column for client type
unique_cust['Client_type'] = ""
```

```python
# fill client type for preferred customers (percentil over 0.75)
unique_cust['Client_type'] =np.where(unique_cust['amount_spent']>=1661.640, 'Preferred_customer','')
unique_cust.head()
```

```python
#same for VIP clients, but by leaving 'preferred customer' in place if condition fails 
unique_cust['Client_type'] =np.where(unique_cust['amount_spent']>=5840.182, 'VIP', unique_cust['Client_type'])
unique_cust.head()
```

```python
unique_cust['Client_type'].value_counts()
#count is the same as number of rows from previous code so seems ok 
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
#adding column with countries to our dataframe
unique_cust['Country']=""
unique_cust['Country']=unique_cust['CustomerID'].map(orders['Country'])
unique_cust['Country'].value_counts()
```

```python
#isolating VIP clients to find most interesting countries
VIP=unique_cust[unique_cust['Client_type']== 'VIP']
VIP['Country'].value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
Pref=unique_cust[unique_cust['Client_type']== 'Preferred_customer']

all_best = [VIP, Pref]
all_best2 = pd.concat(all_best)
all_best2['Country'].value_counts()
```
