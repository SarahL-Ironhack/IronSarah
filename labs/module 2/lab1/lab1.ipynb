```python
# Average steps
steps_avg = fitbit[["Months_encoded", "Steps"]].groupby(["Months_encoded"]).agg("mean") #preparing new DF that will have aggregated data
steps_avg.plot.bar()
plt.savefig("average_steps_months.jpg")

# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_med_months = fitbit[["Months_encoded", "Steps"]].groupby(["Months_encoded"]).agg("median")
steps_med_months.plot.bar()
plt.savefig("median_steps_months.jpg")


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_med_days = fitbit[["Work_or_Weekend", "Steps"]].groupby(["Work_or_Weekend"]).agg("median")
steps_med_days.plot.barh()
plt.savefig("work_or_weekend.jpg")
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
corr = titanic.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))# Generate a mask for the upper triangle

f, ax = plt.subplots(figsize=(11, 9))# Set up the matplotlib figure

cmap = sns.diverging_palette(230, 20, as_cmap=True) # # Generate a custom diverging colormap

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, ### Draw the heatmap with the mask and correct aspect ratio
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
