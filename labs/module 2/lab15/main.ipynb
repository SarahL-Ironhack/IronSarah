```python
#Permutations : nPm   =  n!/(n  m)!
# n = number of objects (8)
#m = number of possibilities (8)
text = 'Ironhack'
lst =[]
for i in text:
    lst.append(i)
print(lst)

def permutations(start, end=[]):
    if len(start) == 0:
        print(end)
    else:
        for i in range(len(start)):
            permutations(start[:i] + start[i+1:], end + start[i:i+1])
permutations(lst)
```

* Challenge 4: Given an array of distinct integers, return all the possible permutations. You can return the answer in any order. The lenght of array and the integers are given by user and may vary

```python
num = str(input('Please enter any integer'))
array = []
for i in num:
    array.append(i)
print(array)

def permutations(start, end=[]):
    if len(start) == 0:
        print(end)
    else:
        for i in range(len(start)):
            permutations(start[:i] + start[i+1:], end + start[i:i+1])
permutations(array)
```
