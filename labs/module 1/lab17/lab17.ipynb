```python
Table_result= wb.data.DataFrame('EG.FEC.RNEW.ZS', ['CAN','USA','DEU','FRA','GBR','JPN','RUS','ITA'],time=range(2014,2016))

result = Table_result.to_json(orient="split")
parsed = json.loads(result)

years=parsed['columns']
countries=parsed['index']
values=parsed['data']


i=0
indexc=0
for c in countries:
    for n in values[i]:
        list_array=(c,years[indexc],n)
        arrays.append(list_array)
        indexc+=1
    indexc=0
    i+=1

final =pd.DataFrame(arrays, columns = ['Country','Year','Value'])
print(final.to_string(index = False))

final.to_csv('Import_table.csv', encoding='utf-8')
```
