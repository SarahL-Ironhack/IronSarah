```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#PART 1 : Upload the data
laburl = (r'C:\Users\Sarah\Downloads\example_data_cleaning.xlsx')
laburl_ori = pd.read_excel(laburl)
lab = laburl_ori.copy()
lab
```

```python
#PART2 : Analyze your data and create a plan for data preparation
#from viewing the DF, column Department has redundant information
# some missing values
# Profession data needs to me harmonized (eg.: Student, Etudiant)
```

```python
lab.info()
```

```python
# Drop column Department
lab.drop('Department', axis=1, inplace=True)
```

```python
lab
```

```python
# view NA data : 2 columns to check, BirthYear and Profession 
lab.isna().sum()
```

```python
#since ther are the same client, replace manually the NaN value by the right one in a cell from the same client
lab['BirthYear'][7]=lab['BirthYear'][6]
lab['BirthYear'][8]=lab['BirthYear'][9]
```

```python
# Same for profession 
lab['Profession'][28]=lab['Profession'][27]
lab['Profession'][29]=lab['Profession'][27]
```

```python
# Harmonize all categorical data for future encoding
lab['Profession'].replace('student','Student',inplace=True)
lab['Profession'].replace('etudient','Student', inplace=True)
lab['Profession'].replace('hr','HR',inplace=True)
lab['Profession'].replace('manager','Manager',inplace=True)
lab['Profession'].replace('bdm','BDM',inplace=True)

#sort Values tu quickly check i didn't forget anything (more complicated with 10000 rows)
lab.sort_values(by=['Profession'])
lab.sort_values(by=['Risk'])
```
